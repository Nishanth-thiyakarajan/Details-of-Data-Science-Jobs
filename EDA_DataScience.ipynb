Creating the Histogram to check the Outliers for Rating.



```python
plt.figure(figsize=(10,8))
plt.ylabel('Rating_Frequency',size=20)
plt.hist(np.array(df['Rating']))
plt.title('Histogram_for_Rating',size=20, color='black')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

Visualising the outlier in box plot.

```python
sns.boxplot(df['Rating'])
plt.title('Box Plot for Rating', size=20, color='black')
plt.show()
```

-1 is the Outlier for Rating.
So we are replacing the outlier with the mean of the rating.

```python
avg=df.loc[df['Rating'] != -1, 'Rating'].mean()
df['Rating']= np.where(df['Rating'] == -1, avg, df['Rating'])
df['Rating']
```

The Outlier was removed in the Rating. To verify, we are uing the density plot from seaborn linbrary. 

```python
plt.figure(figsize=(8,5))
plt.title('Distribution of Rating Column', size=20, color='black')
plt.xlabel('Rating', fontsize=20, color='black')
plt.ylabel('Density', fontsize=20, color='black')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
sns.distplot(df.Rating,color="Red")
plt.show()
```

Top 10 Job locations which has job openings for Data Science roles.

```python
df['Job Location'].unique()
```

```python
df1=df.groupby('Job Location')['Job Title'].count()
df1=df1.reset_index()
df1.rename(columns={'Job Title':'Total Count'}, inplace=True)
df1.sort_values(by=['Total Count'], ascending=False, inplace=True)
df1.reset_index(drop=True, inplace=True)

states_jobs=df1[0:10]
# states_jobs

y_pos=np.arange(len(states_jobs['Job Location']))
plt.figure(figsize=(10,9))
sns.barplot(y_pos,states_jobs['Total Count'])
plt.xticks(y_pos, states_jobs['Job Location'])
plt.xlabel('States', size=20)
plt.ylabel('No of Jobs', size=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title('Top 10 Jobs by State', size=20, color='black')
plt.show() 
```

Top 10 Sectors which has job openings for Data Science roles.

```python
df2=df.groupby('Sector')['Job Title'].count()
df2=df2.reset_index()
df2.rename(columns={'Job Title':'Total Count'}, inplace=True)
df2.sort_values(by=['Total Count'], ascending=False, inplace=True)
df2.reset_index(drop=True, inplace=True)

Sector_jobs=df2[0:10]
# Sector_jobs
y_pos=np.arange(len(Sector_jobs['Sector']))
plt.figure(figsize=(10,9))
sns.barplot(y_pos,Sector_jobs['Total Count'])
plt.xticks(y_pos, Sector_jobs['Sector'],rotation='vertical')
plt.xlabel('Sectors', size=20)
plt.ylabel('No of Jobs', size=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
plt.title('Top 10 Jobs by Sectors', size=20, color='black')
plt.show()
```

Top 10 Industries which has job openings for Data Science roles.

```python
df3=df.groupby('Industry')['Job Title'].count()
df3=df3.reset_index()
df3.rename(columns={'Job Title':'Total Count'}, inplace=True)
df3.sort_values(by=['Total Count'], ascending=False, inplace=True)
df3.reset_index(drop=True, inplace=True)

Industry_jobs=df3[0:10]
# Industry_jobs
y_pos=np.arange(len(Industry_jobs['Industry']))
plt.figure(figsize=(10,9))
sns.barplot(y_pos,Industry_jobs['Total Count'])
plt.xticks(y_pos, Industry_jobs['Industry'],rotation='vertical')
plt.xlabel('Industry', size=20)
plt.ylabel('No of Jobs', size=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
plt.title('Top 10 Jobs by Industry', size=20, color='black')
plt.show()
```

Minimum/Maximum/Average Salaries with respect to Job locations

```python
df4=df[['Job Location', 'Lower Salary', 'Upper Salary', 'Avg Salary(K)']]
df4_summary=df4.groupby('Job Location').agg({'Lower Salary':'mean','Upper Salary':'mean', 'Avg Salary(K)':'mean'})
df4_summary.reset_index(drop=False, inplace=True)
if(states_jobs.shape != df.shape):
    df4_summary=pd.merge(df4_summary, states_jobs[['Job Location']], on='Job Location', how="inner")

width=0.3
ind=np.arange(len(df4_summary['Job Location']))
plt.figure(figsize=(20,10))
bar1 = plt.bar(ind, df4_summary['Lower Salary'], width, color = 'r')
bar2 = plt.bar(ind+width, df4_summary['Upper Salary'], width, color='g')
bar3 = plt.bar(ind+width*2, df4_summary['Avg Salary(K)'], width, color='b')

plt.xlabel("Job Location - States", size=20)
plt.ylabel('Salaries', size=20)
plt.title("Minimum/Maximum/Average for Top 10 Locations", size=20, color='black')
plt.xticks(ind+width, df4_summary['Job Location'],fontsize=20)
plt.yticks(fontsize=20)
plt.legend( (bar1, bar2, bar3), ('Lower Salary', 'Upper Salary', 'Average Salary(K)'),fontsize=15)

plt.show()
```

**Skills Analysis for data scientist.**

Getting the Informations of the Dataframes.

```python
df.info()
```

Skills are added in the form of Boolean values in the dataset.

Extracting the skills to the new dataframe.

```python
df_skills = df.loc[:, 'Python':'google_an']
skills = df_skills.columns.tolist()
print(skills)
```

Visualizing the skills based on the demand in the market.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df_skills.sum().sort_values(ascending=False).values,y=df_skills.sum().sort_values(ascending=False).index)
plt.ylabel('Skills')
plt.xlabel('Apperances in job offers')
plt.title('Most Demanded Skills')
plt.show()
```

We can see that Pyhton, SQL, Excel is the most demanding skills for the data scientist.

Visualizing the skills that are in demand with respect to sectors.

```python
df_sector = df.copy()
df_sector = df_sector[df_sector['Sector'] != '-1']
df_sector = df_sector.groupby('Sector').sum().sort_values(by='Python', ascending=False)
df_sector.head()

fig, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(26, 10))

ax0.bar(df_sector.index, df_sector.Python)
ax0.tick_params(axis='x', labelrotation=90)
ax0.title.set_text('Python')
ax0.set_ylabel('Industries Demanding Python', size=20)

df_sector = df_sector.groupby('Sector').sum().sort_values(by='sql', ascending=False)
ax1.bar(df_sector.index, df_sector.sql)
ax1.tick_params(axis='x', labelrotation=90)
ax1.title.set_text('SQL')
ax1.set_xlabel('Sectors', size=20)
ax1.set_ylabel('Industries Demanding SQL', size=20)

df_sector = df_sector.groupby('Sector').sum().sort_values(by='excel', ascending=False)
ax2.bar(df_sector.index, df_sector.excel)
ax2.tick_params(axis='x', labelrotation=90)
ax2.title.set_text('Excel')
ax2.set_ylabel('Industries Demanding Excel', size=20)

plt.show()
```
